```python
import pandas as pd
from datetime import datetime
from dateutil.parser import parse
from pixiedust.display import *
df3=pd.read_csv('t3.csv')
df1=pd.read_csv('t3 (1).csv')
df1.columns=['Datetime', 'distance','temperature','humidity']
df2=pd.read_csv('t3 (2).csv')
df2.columns=['Datetime', 'distance','temperature','humidity']
result=[df3, df1, df2]
df = pd.concat(result, ignore_index=True)
import numpy as np
date_input= df.Datetime
#datetimeobject = datetime.strptime(date_input,'%Y%m%d')
#new_format_1 = datetimeobject.strftime('%d/%m/%Y')
#df.replace(to_replace = date_input, value = new_format_1)

def parse(date_to_convert):
     return parse('date_to_convert')

df['new_date'] = date_input.apply(parse)

```
